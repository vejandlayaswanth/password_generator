```python
from tkinter import *
import random, string
import pyperclip

# Create main application window
app = Tk()
app.geometry("400x400")
app.resizable(0, 0)
app.title("Python Project -- Password Generator")

# Title labels
Label(app, text='PASSWORD GENERATOR', font='arial 15 bold').pack()
Label(app, text='Python', font='arial 15 bold').pack(side=BOTTOM)

# Password Length Label + Spinbox
Label(app, text='PASSWORD LENGTH', font='arial 10 bold').pack()

# IntVar with default value (important!)
pass_len = IntVar(value=8)

# Spinbox to select length
Spinbox(app, from_=8, to_=32, textvariable=pass_len, width=15).pack()

# StringVar to store generated password
pass_str = StringVar()

# Password Generator Function
def generator():
    password = []

    length = pass_len.get()
    characters = string.ascii_letters + string.digits + string.punctuation

    if length >= 4:
        # Ensure at least one char of each type
        password.append(random.choice(string.ascii_uppercase))
        password.append(random.choice(string.ascii_lowercase))
        password.append(random.choice(string.digits))
        password.append(random.choice(string.punctuation))

        # Fill the rest randomly
        for _ in range(length - 4):
            password.append(random.choice(characters))

        random.shuffle(password)
    else:
        # If length is less than 4, generate directly
        for _ in range(length):
            password.append(random.choice(characters))

    # Join and show password
    pass_str.set(''.join(password))

# Copy to clipboard function
def Copy_password():
    pyperclip.copy(pass_str.get())

# Buttons and Entry
Button(app, text='GENERATE PASSWORD', command=generator).pack(pady=5)
Entry(app, textvariable=pass_str, width=30).pack()
Button(app, text='COPY TO CLIPBOARD', command=Copy_password).pack(pady=5)

# Run the app
app.mainloop()
```
